```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')
```

```python
d = pd.read_excel('C:/Users/abishek/Documents/Abishek_DATA SCIENCE/Python Progess/knn/student_dataset.xlsx')
```

```python
data = d.drop(['S.NO','TOTAL'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
test_set = LabelEncoder().fit_transform(data.iloc[:,3])
```

```python
train_set = data.iloc[:,:-1]
```

```python
train_data = pd.DataFrame(train_set,columns = ['ATTN','THEORY','Daily assn'])
```

```python
test_data = pd.DataFrame(test_set,columns = ['certificate'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
i_train,i_test,o_train,o_test = train_test_split(train_data,test_data,test_size = 0.2,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier as k
```

```python
knn = k(n_neighbors = 5)
knn.fit(i_train,o_train)
```

```python
y_pred = knn.predict(i_test)
```

```python
from sklearn.metrics import confusion_matrix as cm
```

```python
con = cm(o_test,y_pred)
print(con)
```
